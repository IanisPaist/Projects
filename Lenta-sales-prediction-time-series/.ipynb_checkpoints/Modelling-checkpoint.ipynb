```python
import pandas as pd
import numpy as np
import math

import seaborn as sns
sns.set_palette("pastel")

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn.metrics import make_scorer
import catboost as cb
from imblearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import cross_validate, GridSearchCV, cross_val_score, BaseCrossValidator

import json
import warnings
```

```python
def graph_pred_test(df, title, palette='muted'):
    
    #create a pivot table groupped by date (row)
    
    sns.set_style("dark")
    
    #groupby date
    data = pd.pivot_table(data=df,
                  index='date',
                  values=['target', 'pred'],
                  aggfunc='sum').sort_values(by='date')
    
    #calculate error
    data_error = data.copy()
    data_error['diff'] = data_error['target'] - data_error['pred']
    data_error = data_error.drop(columns=['pred', 'target'])
    
    #----------------------------------------------------------------------
    #create figure object
    fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(12,4), sharex=True)
    
    #plot
    sns.lineplot(data=data, palette=palette, ax=axs[0])
    sns.lineplot(data=data_error, ax=axs[1])
    
#     axs[0].set_ylabel('Объем продаж')
#     axs[1].set_ylabel('Объем продаж')
#     axs[0].set_xlabel('Дата')
#     axs[1].set_xlabel('Дата')
    
    plt.suptitle(title)
    axs[0].set_title("Predictions vs Target")
    axs[1].set_title("Error")
    
#     fig.text(0.5,-0.03, "Date", ha="center", va="center", fontsize=15)
    fig.text(0.04, 0.5, 'Сумма продаж', va='center', rotation='vertical', fontsize=12)
    
    axs[0].legend(fontsize=11, loc='upper right')
    axs[1].legend(fontsize=11, loc='upper right')

    plt.show()
```

```python
def get_train_test(df, cv_dict):
    
    #get corresponding indexes
    train_valid_idx = cv_dict['train_1']\
                    + cv_dict['train_2']\
                    + cv_dict['train_3']\
                    + cv_dict['valid_1']\
                    + cv_dict['valid_2']\
                    + cv_dict['valid_3']
    
    test_idx = cv_dict['test']
    
    #select corresponding rows
    X_train = df[df.index.isin(train_valid_idx)]
    X_test = df[df.index.isin(test_idx)]

    #distinguish target from features
    y_train = X_train['target']
    X_train = X_train.drop(columns=['target'])
    
    y_test = X_test['target']
    X_test = X_test.drop(columns=['target'])

    return X_train, y_train, X_test, y_test
```

## Custom функции для Cross validation и WAPE

Создадим custom cross validation для Catboost: три итерации выбора по 14 дней для прогноза в качестве валидационных данных.

Объект валидации генерирует индексы для выбора данных из полного датасета (на основании дат). Индексы были получены при сохранении файлов после препроцессинга в книге EDA and Preprocessing

```python
# class CustomCrossValidation():
    
#     def __init__(self, dict_cv):
#         self.dict_cv = dict_cv
    
#     def split(self, X=None, y=None, groups=None):
#         for i in range(1, 4):
#             training_indices = self.dict_cv[f'train_{i}']
#             valid_indices = self.dict_cv[f'valid_{i}']
            
#             yield training_indices, valid_indices
```

```python
class CustomCrossValidation():
    
    def __init__(self, dict_cv):
        self.dict_cv = dict_cv
    
    def split(self, X=None, y=None, groups=None):
        
        #reset index for X but keep the original index
        X = X.reset_index()
        
        #select train and validation indexes:
        #catboost needs indexes of the simple python list
        #it ignores indexes as per pandas
        #that is why there is a possibility to try to select an index (as per pandas data)
        #outside of the resulted list when we take a slice of the dataframe

        #example: dataframe has 10 rows, indexes from 0 to 9
        #we take a slice of 9th and 10th rows, indexes are kept as 8 as 9
        #but the list created has 2 rows (indexes 0 and 1) - so 8 and 9 are outside
        #so we reset index and search by it as a key to find new indexes
        for i in range(1,4):
            train_idx = X[X['index'].isin(self.dict_cv[f'train_{i}'])].index
            valid_idx = X[X['index'].isin(self.dict_cv[f'valid_{i}'])].index
        
            yield train_idx, valid_idx
```

Создадим метрику

```python
#функция WAPE
def wape(y_true: np.array, y_pred: np.array):
    return np.sum(np.abs(y_true-y_pred))/np.sum(np.abs(y_true))
```

WAPE в формате catboost

```python
#функция WAPE для catboost
class custom_wape:
    
    def get_final_error(self,error,weight):
        return error
    
    def is_max_optimal(self):
        return False
    
    def evaluate(self, approxes, target, weight):
        
        assert len(approxes) == 1
        assert len(target) == len(approxes[0])
        
        y_true = target
        y_pred = approxes[0]
        
        # Check for NaN or infinite values
        if np.any(np.isnan(y_true)) or np.any(np.isnan(y_pred)) or np.any(np.isinf(y_true)) or np.any(np.isinf(y_pred)):
            raise ValueError("NaN or infinite values in y_true or y_pred")
        
        return np.sum(np.abs(y_true - y_pred)) / np.sum(np.sum(np.abs(y_true)) + 0.00000000001) , 0.0
```

```python
%%time

#if needed to launch, choose 1. To skip = 0
start = 0 

if start == 1:
    
    #define CatBoost model
    model = cb.CatBoostRegressor(random_state=99,
                                 eval_metric=custom_wape(),
                                 iterations=500,
                                 early_stopping_rounds=15)

    #create catboost Pool object
    ds = cb.Pool(X_train, label=y_train, cat_features=features_str)

    #dict with hyperparameters
    hp = {'depth': [3, 6, 9],
            'l2_leaf_reg': [1, 3, 5]}

    #create instance of cross validation
    custom_cv = CustomCrossValidation(dict_cv=df_dict).split(X=X_train)

    #lauch grid search
    grid_search_result  = model.grid_search(param_grid=hp,
                                            X=ds,
                                            cv=custom_cv,
                                            verbose=True)

    #SAVE THE BEST MODEL
    model.save_model(PATH_TO_MODEL_FOLDER + NAME_MODEL)
    
    #print best parameters
    grid_search_result['params']
```

```python
X_train, y_train, X_test, y_test = get_train_test(df, df_dict)
```

```python
print('X_train.shape:', X_train.shape)
print('X_test.shape', X_test.shape)

print('y_train.shape:', y_train.shape)
print('y_test.shape', y_test.shape)
```

Выделим фичи категориальные и числовые (категориальные далее понадобятся для Catboost)

```python
%%capture

#create an instance of scaling and encoding the features

#data transformation objects
#using OE (not OHE) for tree-like models (incl boosting) because two much unuique values -> 
#to many transformed columns -> longer time and OHE for linear regression
# sc = StandardScaler()
# oe = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
# ohe = OneHotEncoder(handle_unknown='ignore')

#list of numeric and non-numeric features
    #days_ny, days_paskha - not encoding      

features_n = ['day',
            'dayofweek',
            'ma_3d',
            'ma_6d',
            'ma_9d',
            'ma_12d',
            'ma_15d',
            'max_3d',
            'max_6d',
            'max_9d',
            'max_12d',
            'max_15d',
            'min_3d',
            'min_6d',
            'min_9d',
            'min_12d',
            'min_15d',
            'sales_0days_ago',
            'sales_1days_ago',
             'sales_2days_ago',
             'sales_3days_ago',
             'sales_4days_ago',
             'sales_5days_ago',
             'sales_6days_ago',
             'sales_7days_ago',
             'sales_8days_ago',
             'sales_9days_ago',
             'sales_10days_ago',
             'sales_11days_ago',
             'sales_12days_ago',
             'sales_13days_ago',
             'sales_14days_ago',
             'sales_15days_ago',
             'sales_16days_ago',
             'sales_17days_ago',
             'sales_18days_ago',
             'sales_19days_ago',
             'sales_20days_ago',
             'sales_21days_ago']

features_str = ['st_id',
                'pr_sku_id',
                'st_city_id',
                'st_division_code',
                'st_type_format_id',
                'st_type_loc_id',
                'st_type_size_id',
                'pr_group_id',
                'pr_cat_id',
                'pr_subcat_id',
                'pr_uom_id',
                'holiday',
               ]

#change all categorical to string
for col in X_train[features_str].columns:
    X_train[col] = X_train[col].astype(str)
```

####  Обучение модели

Обучим модель, перебирая различные гиперпараметры:

```python
%%time

#if needed to launch, choose 1. To skip = 0
start = 0 

if start == 1:
    
    #define CatBoost model
    model = cb.CatBoostRegressor(random_state=99,
                                 eval_metric=custom_wape(),
                                 iterations=500,
                                 early_stopping_rounds=15)

    #create catboost Pool object
    ds = cb.Pool(X_train, label=y_train, cat_features=features_str)

    #dict with hyperparameters
    hp = {'depth': [3, 6, 9],
        'l2_leaf_reg': [1, 3, 5]}

    #create instance of cross validation
    custom_cv = CustomCrossValidation(dict_cv=df_dict).split(X=X_train)

    #lauch grid search
    grid_search_result  = model.grid_search(param_grid=hp,
                                            X=ds,
                                            cv=custom_cv,
                                            verbose=True)

    #SAVE THE BEST MODEL
    model.save_model(PATH_TO_MODEL_FOLDER + NAME_MODEL)
    
    #print best parameters
    grid_search_result['params']
```

Наилучшие параметры получились: 

* {'depth': 9, 'l2_leaf_reg': 3}

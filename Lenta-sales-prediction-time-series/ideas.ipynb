Ответы на вопросы для 8-й и 14-й команд:
1. Общую сумму продаж в рублях мы не используем при обучении, поскольку данный признак непосредственно зависит от значения целевого признака. Мы можем на основе суммы и кол-ва сгененировать отдельный признак / временной ряд с ценой товара, обучить на нем ARIMA модель или Facebook Prophet и сделать предсказания на 2 недели вперед для теста. Таким образом мы сможем получить на тесте общую сумму продаж, переменожив предсказания основной модели (кол-во) с предсказанной ценой. Данный прогноз может понадобиться в финальном решении.
2. По поводу использования корреляции между магазинами или товарами - ее можно использовать для класстеризации товаров/магазинов.
